```python
from sklearn.cluster import KMeans 

X1 = data[['Annual Income (k$)','Spending Score (1-100)']].values
inercia = []
for n in range(1,11):
    algoritmo = KMeans(n_clusters = n)
    algoritmo.fit(X1)
    inercia.append(algoritmo.inertia_)
```

```python
plt.figure(figsize = (15,6))
plt.plot(np.arange(1,11), inercia,'o')
plt.plot(np.arange(1,11), inercia,'-',alpha = 0.5)
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.title('Metodo del codo')
plt.show()
```

```python
algoritmo = KMeans(n_clusters = 5)
algoritmo.fit(X1)
labels1 = algoritmo.labels_
centroides = algoritmo.cluster_centers_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(12,4))
ax1.scatter(X1[:][:,0],X1[:][:,1],c=labels1)
ax1.scatter(centroides[:,0],centroides[:,1],s=50,c='r')
ax1.set_title('Algoritmo de K - Medias')

ax2.scatter(X1[:][:,0],X1[:][:,1])
ax2.set_title('Datos originales')
```
